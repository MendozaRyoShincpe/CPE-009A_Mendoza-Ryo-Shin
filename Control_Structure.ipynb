**1. If-Statement**

```python
a = 12
b = 100
if b > a:
  print("b is greater than a")
```

**2. Elif Statement**

```python
a = 12
b = 12
if b > a:
  print("b is greater than a")
elif a == b:
  print ("a and b are equal")
```

**3. Statement Else-If**

```python
a = 100
b = 12

if b > a:
  print("b is greater than a")
elif a == b:
  print ("a and b are equal")
else:
  print("a is greater than b")
```

**4. Short Hand If Statement**

```python
if a > b: print ("a is greater thab b")
```

**5. Short Hand If.. Else Statement**

```python
a = 2
b = 330
print("A") if a > b else print ("B")
```

**6. And-Both conditions are true**

```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print ("Both conditions are true")
```

**7. Or-atleast one of the conditions is true**

```python
a = 200
b = 33
c = 500
if a > b or a > c:
  print ("At least one of the conditions is true")
```
