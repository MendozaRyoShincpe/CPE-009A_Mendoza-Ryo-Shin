**Procedure:**

A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

1. PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)

2. PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment

3. MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))

4. MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment

5. FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))

6. FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment

7. HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.

8. Major exams are inputted out of 100%.

9. Show the codes that successfully run the program.

10. Provide comments or documentation strings for your program.

```python
print("----Prelim----")
PreHOA = float(input("Enter Prelim Hands-on Activity grade: "))
PreQUIZ = float(input("Enter Prelim Quiz grade: "))
PreAS = float(input("Enter Prelim Assignment grade: "))

PreCS = PreHOA*.5 + PreQUIZ*.3 + PreAS*.2
print("Prelim CS is: ", PreCS)

PreEX = float(input("Enter Prelim Exam grade: "))
PG = PreEX*.5 + PreCS*.5
print("Prelim Grade is: ", "%.2f" % PG)

print("\n----Midterm----")
MidHOA = float(input("Enter Midterm Hands-on Activity grade: "))
MidQUIZ = float(input("Enter Prelim Quiz grade: "))
MidAS = float(input("Enter Prelim Assignment grade: "))

MidCS = MidHOA*.5 + MidQUIZ*.3 + MidAS*.2
print("Midterm CS is: ", MidCS)

MidEX = float(input("Enter Prelim Exam grade: "))
MG = PG*.3 + (MidEX*.5 + MidCS*.5)*.7
print("Midterm Grade is: ", "%.2f" % MG)

print("\n----Finals----")
FinHOA = float(input("Enter Finals Hands-on Activity grade: "))
FinQUIZ = float(input("Enter Prelim Quiz grade: "))
FinAS = float(input("Enter Prelim Assignment grade: "))

FinCS = FinHOA*.5 + FinQUIZ*.3 + FinAS*.2
print("Final CS is: ", FinCS)

FinEX = float(input("Enter Prelim Exam grade: "))
FG = MG*.3 + (FinEX*.5 + FinCS*.5)*.7
print("Final Grade is: ", "%.2f" % FG)

```

```python
# Ensuring 3 students will enter the program (1)
for student in range(1, 4): #It will loop 3 times, 1 is the starting number, 4 is the ending number, which is not included
    print("\n==============================")
    name = input(f"Enter name of the Student {student}: ") #It will indicate as the number of student counts
    print("==============================")

    # -------- PRELIM --------
    print("\n---- Prelim ----")
    PreHOA = float(input("Enter Prelim Hands-on Activity grade: "))
    PreQUIZ = float(input("Enter Prelim Quiz grade: "))
    PreAS = float(input("Enter Prelim Assignment grade: "))

    PreCS = PreHOA * 0.5 + PreQUIZ * 0.3 + PreAS * 0.2
    print("Prelim CS is:", PreCS)

    PreEX = float(input("Enter Prelim Exam grade: "))
    PG = PreEX * 0.5 + PreCS * 0.5 # Formula to get the Prelim Grade

    # -------- MIDTERM --------
    print("\n---- Midterm ----")
    MidHOA = float(input("Enter Midterm Hands-on Activity grade: "))
    MidQUIZ = float(input("Enter Midterm Quiz grade: "))
    MidAS = float(input("Enter Midterm Assignment grade: "))

    MidCS = MidHOA * 0.5 + MidQUIZ * 0.3 + MidAS * 0.2
    print("Midterm CS is:", MidCS)

    MidEX = float(input("Enter Midterm Exam grade: "))
    MG = PG * (1/3) + (MidEX * 0.5 + MidCS * 0.5) * (2/3) # Formula to get the Midterm Grade

    # -------- FINALS --------
    print("\n---- Finals ----")
    FinHOA = float(input("Enter Finals Hands-on Activity grade: "))
    FinQUIZ = float(input("Enter Finals Quiz grade: "))
    FinAS = float(input("Enter Finals Assignment grade: "))

    FinCS = FinHOA * 0.5 + FinQUIZ * 0.3 + FinAS * 0.2
    print("Final CS is:", FinCS)

    FinEX = float(input("Enter Finals Exam grade: "))
    FG = MG * (1/3) + (FinEX * 0.5 + FinCS * 0.5) * (2/3) # Formula to get the Final grade

    # Showing the inputted name of the student(2)
    print("\n===Student Grade===")
    print("Name: ", name)
    print("Prelim Grade is:", "%.2f" % PG)
    print("Midterm Grade is:", "%.2f" % MG)
    print("Final Grade is:", "%.2f" % FG)

    # Conditional statements to the Numerical Grade(3)
    if FG >= 99 and FG <= 100:
        print("Numerical Grade is: 1.00")
    elif FG >= 96 and FG <= 98.99:
        print("Numerical Grade is: 1.25")
    elif FG >= 93 and FG <= 95.99:
       print("Numerical Grade is: 1.50")
    elif FG >= 90 and FG <= 92.99:
       print("Numerical Grade is: 1.75")
    elif FG >= 87 and FG <= 89.99:
       print("Numerical Grade is: 2.00")
    elif FG >= 84 and FG <= 86.99:
       print("Numerical Grade is: 2.25")
    elif FG >= 81 and FG <= 83.99:
       print("Numerical Grade is: 2.50")
    elif FG >= 78 and FG <= 80.99:
       print("Numerical Grade is: 2.75")
    elif FG >= 75 and FG <= 77.99:
       print("Numerical Grade is: 3.00")
    else:
       print("Failed")
```
