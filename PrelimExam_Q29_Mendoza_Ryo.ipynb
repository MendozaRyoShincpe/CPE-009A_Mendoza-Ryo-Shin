```python
subject_name = str(input("Enter Subject: "))
student_name = str(input("Enter Student Name: "))
grade = float(input("Enter grade percentage: "))

print("\n----------Grade Remarks---------")
print("Student Name:", student_name)
print("Subject Name:", subject_name)
print("Grade Percentage:", grade,"%\n")

if grade >= 90 and grade <= 100:
    marks = "A+"
    print("You are Outstanding!")
    print("Mark:", marks)
elif grade >= 80 and grade <= 89:
    marks = "A"
    print("You are Excellent!")
    print("Mark:", marks)
elif grade >= 70 and grade <= 79:
   marks = "B+"
   print("You are Very Good!")
   print("Mark:", marks)
elif grade >= 60 and grade <= 69:
    marks = "B"
    print("You are Good!")
    print("Mark:", marks)
elif grade >= 50 and grade <= 59:
    marks = "C"
    print("You are Satisfactory!")
    print("Mark:", marks)
elif grade >= 40 and grade <= 49:
    marks = "D"
    print("You are Needs Improvement!")
    print("Mark:", marks)
elif grade >= 0 and grade <= 39:
    marks = "F"
    print("You are Fail!")
    print("Mark:", marks)
else:
    print("Invalid input!") # This will print if the user choose between less 0, or greater than 100
```
